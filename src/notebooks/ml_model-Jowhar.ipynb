```python
data_path = "data/interim/ingestion/"
filename = "Preprocessor_001-f7-Jowhar.csv"

config_path = "config/"
config_filename = "config.ini"

station_name = "Jowhar"
```

```python
import pandas as pd

df = pd.read_csv(data_path + filename)
df.head()
```

```python
df = df.drop(columns="location").set_index("date")
df.head()
```

```python
# Identify local peaks in the data
from scipy.signal import find_peaks
peaks, _ = find_peaks(
    df['level__m'] + df["y"], 
    distance=10, 
    height=(4.5, None),  # ignore peaks below 4.5m
    prominence=0.8
)
df['is_peak'] = False
df.loc[df.index[peaks], 'is_peak'] = True

# load predictions
df_pred = pd.read_csv("data/interim/evaluation/Preprocessor_001-f7-Jowhar-RandomForestRegressor_001-predictions.csv")
df_pred["date"] = pd.to_datetime(df_pred["date"])
df_pred = df_pred.set_index("date")


import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Ensure the index is datetime
df.index = pd.to_datetime(df.index)

plt.figure(figsize=(15,5), clear=True)
for peak in peaks:
    plt.axvline(x=df.index[peak], color='r', linestyle='--')
plt.plot(df_pred.index, df_pred['abs_pred_y'], color='deepskyblue', label='Predicted Water Level + y')
plt.plot(df.index, df['level__m'] + df["y"], color='darkgreen', label='Water Level + y')
plt.plot(df.index[peaks], (df['level__m'] + df['y']).iloc[peaks], "D", color='r', label='Peaks')
plt.title(f'{station_name}: Water Level to Predict over time (with Peaks)')
plt.xlabel('Date')
plt.ylabel('Water Level to Predict (m)')
# set ticks every month (index is datetime)
ax = plt.gca()
ax.set_xlim([df.index.min(), df.index.max()])
ax.set_ylim([0, df['level__m'].max() + df['y'].max() + 1])
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=90)

# add grid with minor ticks (at 0.5m intervals)
ax.yaxis.set_minor_locator(plt.MultipleLocator(0.5))
ax.grid(which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()

print(f"Identified {len(peaks)} peaks:")
for peak_date, peak_value in zip(df.index[peaks].to_list(), (df['level__m'] + df['y']).iloc[peaks].to_list()):
    print(f" - at {peak_date}: {peak_value:.2f} m")

print(df.index.min(), df.index.max())
```

```python
# Generate subcharts for each peak
for peak in peaks:
    start_idx = max(0, peak - 30)
    end_idx = min(len(df), peak + 10)
    
    plt.figure(figsize=(10,4), clear=True)
    plt.plot(df_pred.index, df_pred['abs_pred_y'], color='deepskyblue', label='Predicted Water Level + y')
    plt.plot(df.index[start_idx:end_idx], df['level__m'][start_idx:end_idx] + df['y'][start_idx:end_idx], color='darkgreen', label='Water Level + y')
    plt.plot(df.index[peak], (df['level__m'] + df['y']).iloc[peak], "D", color='r')
    plt.axvline(x=df.index[peak], color='r', linestyle='--', label='Peak')
    plt.title(f'{station_name}: Water Level around Peak on {df.index[peak].strftime("%Y-%m-%d")}')
    plt.xlabel('Date')
    plt.ylabel('Water Level to Predict (m)')
    plt.xticks(rotation=90)
    plt.xlim([df.index[start_idx], df.index[end_idx-1]])
    plt.ylim([0, (df['level__m'] + df['y']).max() + 1])
    ax = plt.gca()
    ax.yaxis.set_minor_locator(plt.MultipleLocator(0.5))
    ax.xaxis.set_major_locator(mdates.DayLocator())
    ax.grid(which='both', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.show()
```

```python
import shap
from flood_forecaster.ml_model.registry import MODEL_MANAGER_REGISTRY


# Split the data into training and test sets
train_df = df[df.index < "2023-10-01"].drop(columns=["is_peak"])
test_df = df[df.index >= "2023-10-01"]

# Filter test data for peak events
# NOTE: considering peak event start date here!
    # Peaks:
    # 2022-04-29 --> 2022-04-22 (small increase) or 23 (major increase)
    # 2022-08-17 --> 2022-08-14
    # 2022-11-19 --> 2022-11-15 (small increase) or 18 (major increase)
    # 2023-03-25 --> 2023-03-19 (small increase) or 22 (major increase) or 23 (critical increase)
    # 2024-04-14 --> 2024-04-06 (small increase) or 08 (major increase) or 09 (critical increase)
    # 2025-04-01 --> 2025-03-29 (first sharp increase) or 01 (critical increase)
    # 2025-04-20 --> 2025-04-17
    # 2025-05-07 --> 2025-05-02 (first sharp increase) or 03 (steady continuation)
test_df['is_peak'] = test_df.index.astype(str).isin([
    "2024-04-06", "2024-04-08", "2024-04-09",
    "2025-03-29", "2025-04-01",
    "2025-04-17",
    "2025-05-02", "2025-05-03"
])
peak_test_df = test_df[test_df['is_peak']]

model = MODEL_MANAGER_REGISTRY["RandomForestRegressor_001"].load(model_path="models/", model_name="Preprocessor_001-f7-RandomForestRegressor_001-Jowhar")
print(model.feature_names_in_)

# # QUICKFIX: remove lagabs features from train_df for SHAP explainer
# from flood_forecaster.ml_model.RandomForestRegressor001.model import filter_columns
# train_df = filter_columns(train_df, remove_y=True, remove_level=True, remove_date=True)
# peak_test_df = filter_columns(peak_test_df, remove_y=True, remove_level=False, remove_date=True)

explainer = shap.Explainer(model, train_df)


# Generate SHAP values for the test data
shap_values = explainer(peak_test_df.drop(columns=["is_peak"]))

# # Plot waterfall charts for each peak event
# for i in range(len(peak_test_df)):
#     shap.plots.waterfall(shap_values[i])

# Plot custom waterfall charts for each peak event
for i in range(len(peak_test_df)):
    # Extract SHAP values and base value for the current instance
    shap_value = shap_values[i]
    base_value = shap_value.base_values
    feature_values = shap_value.data
    shap_contributions = shap_value.values

    # Extract additional information
    peak_date = peak_test_df.index[i]
    expected_value = peak_test_df.iloc[i]["level__m"] + peak_test_df.iloc[i]["y"]
    predicted_value = expected_value + shap_contributions.sum()

    # Create a custom waterfall chart
    plt.figure(figsize=(10, 6))
    shap.waterfall_plot(shap_value, show=False)
    plt.title(f"SHAP Waterfall Chart for Peak on {peak_date}")
    plt.xlabel("Contribution to Prediction")
    plt.figtext(0.5, -0.1, f"Expected Value: {expected_value:.2f} m\nPredicted Value: {predicted_value:.2f} m", 
                wrap=True, horizontalalignment='center', fontsize=10)
    plt.show()
```
